```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_excel("../housepricing/HousePricePrediction.xlsx")
print(data.head(10))
```

```python
print(data.info())
print(data.isnull().sum())
```

```python
print(data.MSZoning.unique())
data["MSZoning"] = data["MSZoning"].map({
    "RL": 0,
    "RM": 1,
    "C (all)": 2,
    "FV": 3,
    "RH": 4
})

print(data.head(10))
```

```python
data.LotConfig.unique()
data["LotConfig"]=data["LotConfig"].map({"Inside":0,"FR2":1,"Corner":2,"CulDSac":3,"FR3":4})
```

```python
data.BldgType.unique()
data["BldgType"]=data["BldgType"].map({'1Fam':0, '2fmCon':1, 'Duplex':2, 'TwnhsE':3, 'Twnhs':4})
```

```python
data.Exterior1st.unique()
data["Exterior1st"]=data["Exterior1st"].map({'VinylSd':0, 'MetalSd':1, 'Wd Sdng':2, 'HdBoard':3, 'BrkFace':4, 'WdShing':5,
       'CemntBd':6, 'Plywood':7, 'AsbShng':8, 'Stucco':9, 'BrkComm':10, 'AsphShn':11,
       'Stone':12, 'ImStucc':13, 'CBlock':14, "nan":"NaN"})
```

```python
data.drop(["Id","BsmtFinSF2"],axis=1)
```

```python
print(data.isnull().sum())
meansaleprice=data["SalePrice"].mean()
data.loc[data.SalePrice.isnull()]=meansaleprice
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
train_X,test_x,train_Y,test_y=train_test_split(
    data.drop(["SalePrice"],axis=1),
    data["SalePrice"],
    test_size=0.2,
    random_state=42
    )
lrmod=LinearRegression()
lrmod.fit(train_X,train_Y)
y_predict=lrmod.predict(test_x)
mae = mean_absolute_error(test_y, y_predict)
mse = mean_squared_error(test_y, y_predict)
r2 = r2_score(test_y, y_predict)

print("MAE:", mae)
print("MSE:", mse)
print("R²:", r2)
```
